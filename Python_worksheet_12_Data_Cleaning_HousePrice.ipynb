## Ans:12    Considering the 'Price' column as target variable which is to be predicted, how will you treat the missing values? Give a brief explanation and treat the missing values in 'Price' column

```python
df1['Price'].isnull().sum()
```

```python
(df1['Price'].isnull().sum()/len(df1))*100
```

We can not drop 21% of null values, so we will treat the missing value by fillna method using mean 

```python
df1['Price'].fillna(df1['Price'].mean(),inplace=True)
```

```python
df1['Price'].isnull().sum()
```

Now we do not have any null values in the Price column.
